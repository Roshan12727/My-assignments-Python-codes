**Exercise 1: Prime Numbers**

```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

**Exercise 2: Product of Random Numbers**

```python
import random

a = random.randint(1, 10)
b = random.randint(1, 10)

answer = int(input(f"What is {a} * {b}? "))

if answer == a * b:
    print("Correct")
else:
    print("Wrong")
    print("Correct answer is", a * b)
```

**Exercise 3: Squares of Even/Odd Numbers**

```python
for num in range(100, 201):
    if num % 2 == 0:
        print(num, "square is", num * num)
```

**Exercise 4:
Problem Statement (Question): Hospital Billing System**

```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

days = 4
lab_tests = 3
medicine_charges = 2400

total = (room_charge_per_day * days) + doctor_fee + (lab_test_charge * lab_tests) + medicine_charges

if total > 10000:
    discount = total * 0.10
    total = total - discount

print("Final Bill Amount: ₹", total)
```
